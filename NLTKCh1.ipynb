```python
# You may want to repeat such calculations on several texts, but it is tedious to keep retyping the 
# formula. Instead, you can come up with your own name for a task, like "lexical_diversity" or 
# "percentage", and associate it with a block of code. Now you only have to type a short name instead of 
# one or more complete lines of Python code, and you can re-use it as often as you like. The block of code 
# that does a task for us is called a function, and we define a short name for our function with the 
# keyword def. The next example shows how to define two new functions, lexical_diversity() and 
# percentage():

def lexical_diversity (text):
    return len(set(text)) / len(text)

# Lexical diversity divides number of word tokens by total number of words and returns this value 

def percentage (count, total):
    return 100 * count / total

# Percentage divides count by total and multiplies by 100, returning this value
```

```python
# Next, let's look at the long words of a text; perhaps these will be more characteristic and informative. 
# For this we adapt some notation from set theory. We would like to find the words from the vocabulary of 
# the text that are more than 15 characters long.

V = set(text1) # V is the vocabulary for text1

long_words = [w for w in V if len(w) > 15] # all the words in the vocabulary with length longer than 15
# In a longer explanation... long_words = some sequence of w
# You only include w if it is a word of longer than 15 characters in V
# Note... You can represent long_words as anything. For example, you could say
# long_words = [2w for w in V if len(w) > 15]

sorted(long_words) # sorting these long words
# also note that the sort starts with capitalized words and then moves on to lowercase words in 
# alphabetical order.
```

```python
# Well, these very long words are often hapaxes (i.e., unique) and perhaps it would be better to find 
# frequently occurring long words. This seems promising since it eliminates frequent short words 
# (e.g., the) and infrequent long words (e.g. antiphilosophists). Here are all words from the chat corpus 
# that are longer than seven characters, that occur more than seven times:

fdist5 = FreqDist(text5)
# fdist5 is the frequency distribution for text5

sorted(w for w in set(text5) if len(w) > 7 and fdist5[w] > 7)
# if each word in the vocabulary of text5 has length greater than 7 AND is counted more than 7 times
# in the frequency distribution of text5, then include this word as part of a sorted list
```

```python
# A collocation is a sequence of words that occur together unusually often. Thus red wine is a 
# collocation, whereas the wine is not. A characteristic of collocations is that they are resistant to 
# substitution with words that have similar senses; for example, maroon wine sounds definitely odd.

# To get a handle on collocations, we start off by extracting from a text a list of word pairs, also known 
# as bigrams. This is easily accomplished with the function bigrams():

list(text4.bigrams(['more', 'is', 'said', 'than', 'done']))

# One way to think about collocations... When companies do "spinning", they try to replace one word for
# another similar word. This MIGHT work but only if the word is not part of a collocation. Maybe this is
# what WordAI does.
```

```python
# Here we see that the pair of words than-done is a bigram, and we write it in Python as ('than', 'done'). 
# Now, collocations are essentially just frequent bigrams, except that we want to pay more attention to 
# the cases that involve rare words. In particular, we want to find bigrams that occur more often than we 
# would expect based on the frequency of the individual words. The collocations() function does this for 
# us. We will see how it works later.

text4.collocations()
```

```python
# Counting words is useful, but we can count other things too. For example, we can look at the 
# distribution of word lengths in a text, by creating a FreqDist out of a long list of numbers, where each 
# number is the length of the corresponding word in the text:

# We start by deriving a list of the lengths of words in text1
[len(w) for w in text1]
```

```python
# FreqDist then counts the number of times each of these occurs 
fdist = FreqDist(len(w) for w in text1)
```

```python
# The result [3] is a distribution containing a quarter of a million items, each of which is a number 
# corresponding to a word token in the text. But there are at most only 20 distinct items being counted, 
# the numbers 1 through 20, because there are only 20 different word lengths. I.e., there are words 
# consisting of just one character, two characters, ..., twenty characters, but none with twenty one or more 
# characters.

print(fdist)
```

```python
# One might wonder how frequent the different lengths of word are (e.g., how many words of length four 
# appear in the text, are there more words of length five than length four, etc). We can do this as 
# follows:

fdist.most_common()
```
